```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Initial network structure without conditional bonuses
network = {
    'Pioneer': {
        'count': 36,
        'personal_sales': 30e6,
        'comm_rate': 0.40,
        'override_sales': 0,
        'override_rate': 0.0,
        'monthly_bonus': 0,
        'annual_bonus': 0
    },
    'Catalyst': {
        'count': 9,
        'personal_sales': 50e6,
        'comm_rate': 0.425,
        'override_sales': 4 * 30e6,
        'override_rate': 0.03,
        'monthly_bonus': 0,  # Will be updated after system_sales calculation
        'annual_bonus': 0    # Will be updated after system_sales calculation
    },
    'Visionary': {
        'count': 3,
        'personal_sales': 100e6,
        'comm_rate': 0.425,
        'override_sales': 3 * 50e6,
        'override_rate': 0.03,
        'monthly_bonus': 0,  # Will be updated after system_sales calculation
        'annual_bonus': 0    # Will be updated after system_sales calculation
    },
    'Trailblazer': {
        'count': 1,
        'personal_sales': 150e6,
        'comm_rate': 0.425,
        'override_sales': 0,  # Will be updated after system_sales calculation
        'override_rate': 0.05,
        'monthly_bonus': 0,   # Will be updated after system_sales calculation
        'annual_bonus': 0     # Will be updated after system_sales calculation
    }
}

# Calculate system sales
system_sales_monthly = sum(params['count'] * params['personal_sales'] for params in network.values())
system_sales_annual = 0

# Update network with calculated values
network['Trailblazer']['override_sales'] = system_sales_monthly
network['Catalyst']['monthly_bonus'] = 10e6 if system_sales_monthly >= 200e6 else 0
network['Catalyst']['annual_bonus'] = 50e6 if (50e6 * 12) >= 3e9 else 0
network['Visionary']['monthly_bonus'] = 20e6 if system_sales_monthly >= 700e6 else 0
network['Visionary']['annual_bonus'] = 100e6 if system_sales_annual >= 5e9 else 0
network['Trailblazer']['monthly_bonus'] = 50e6 if system_sales_monthly >= 2.25e9 else 0
network['Trailblazer']['annual_bonus'] = 200e6 if system_sales_annual >= 30e9 else 0

results = []
for role, params in network.items():
    personal_comm = params['personal_sales'] * params['comm_rate']
    override_comm = params['override_sales'] * params['override_rate']
    total_comm_person = personal_comm + override_comm
    total_comm_all = total_comm_person * params['count']

    monthly_bonus_person = params['monthly_bonus']
    annual_bonus_person = params['annual_bonus']
    total_bonus_all = (monthly_bonus_person + annual_bonus_person) * params['count']

    results.append({
        'Cp ': role,
        'S ngi': params['count'],
        'Doanh s c nhn (VND/thng)': f"{params['personal_sales']:,.0f}",
        'Doanh s h thng (VND/thng)': f"{system_sales_monthly:,.0f}",
        'Hoa hng c nhn / ngi (VND)':  f"{personal_comm:,.0f}",
        'Hoa hng h thng / ngi (VND)': f"{override_comm:,.0f}",
        'Tng hoa hng / ngi (VND)':     f"{total_comm_person:,.0f}",
        'Tng hoa hng (c cp) (VND)':    f"{total_comm_all:,.0f}",
        'Thng thng / ngi (VND)':      f"{monthly_bonus_person:,.0f}",
        'Thng nm / ngi (VND)':        f"{annual_bonus_person:,.0f}",
        'Tng thng (c cp) (VND)':      f"{total_bonus_all:,.0f}"
    })

df_network = pd.DataFrame(results)

# Set up the style
# plt.style.use('seaborn')
plt.figure(figsize=(15, 8))

# Create bar chart
x = df_network['Cp ']
y1 = df_network['Tng hoa hng / ngi (VND)'].str.replace(',', '').astype(float)
y2 = df_network['Tng thng (c cp) (VND)'].str.replace(',', '').astype(float)

# Plot bars
bar_width = 0.35
plt.bar(x, y1/1e6, bar_width, label='Hoa hng/ngi (triu VND)', color='skyblue')
plt.bar([i for i in range(len(x))], y2/1e6, bar_width,
        bottom=y1/1e6, label='Thng/cp (triu VND)', color='lightgreen')

# Customize the chart
plt.title('Phn tch hoa hng v thng theo cp ', pad=20, fontsize=14)
plt.xlabel('Cp ')
plt.ylabel('Triu VND')
plt.legend()
plt.xticks(rotation=45)

# Add value labels on bars
for i, v in enumerate(y1/1e6):
    plt.text(i, v/2, f'{v:,.0f}M', ha='center', color='black')
for i, v in enumerate(y2/1e6):
    if v > 0:
        plt.text(i, v + y1[i]/1e6 - v/2, f'{v:,.0f}M', ha='center', color='black')

plt.tight_layout()
plt.show()

# Calculate and display summary
total_commission = df_network['Tng hoa hng (c cp) (VND)'].str.replace(',', '').astype(float).sum()
total_bonus = df_network['Tng thng (c cp) (VND)'].str.replace(',', '').astype(float).sum()
total_payout = total_commission + total_bonus

print("\nTng kt h thng:")
print(f"Tng hoa hng: {total_commission:,.0f} VND")
print(f"Tng thng: {total_bonus:,.0f} VND")
print(f"Tng chi tr: {total_payout:,.0f} VND")

# Display the detailed table
print("\nBng chi tit:")
print(df_network.to_string(index=False))


```
